```python
c= df['vote_average'].mean()
m= df['vote_count'].quantile(0.2)
def rating(x, m=m, c=c):
    v = x['vote_count']
    r = x['vote_average']
    return (v/(v+m) * r) + (m/(m+v) * c)
```

```python
df['rate'] = df.apply(rating, axis=1)
```

```python
def date(x):
    if x < 2000:
        return 'Old'
    elif 2000 < x < 2010:
        return 'Modern'
    else:
        return 'New'
```

```python
def rate(x):
    if x < 5.5:
        return 'Low'
    elif 5.5 < x < 7.5:
        return 'Medium'
    else:
        return 'High'
```

```python
def runtime(x):
    if x < 80:
        return 'Short'
    elif 80 < x < 120:
        return 'Normal'
    elif 120 < x < 200:
        return 'Long'
    else:
        return 'Very Long'
```

```python
df['runtime']=df['runtime'].apply(lambda x: runtime(x))
df['rate']=df['rate'].apply(lambda x: rate(x))
df['date']=df['date'].apply(lambda x: date(x))

```

# preproccessing

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tf = TfidfVectorizer(analyzer='word', stop_words='english')
tfidf_matrix = tf.fit_transform(df['overview'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python

titles = df['title']
indices = pd.Series(df.index, index=df['title'])
```

```python
def get_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

```python
get_recommendations('Good')
```
